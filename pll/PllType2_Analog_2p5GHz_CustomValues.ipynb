# Installing Jupyter Notebook

Install Jupyter Notebook using pip or conda, and launch it to create a new notebook file.

# Converting a Python Script to a Notebook

Import a Python script into a Jupyter notebook by copying code into cells and separating logic into executable blocks.

# Using Cell Magic Commands

Apply cell magic commands like %%time or %%matplotlib inline to modify code execution and output behavior in cells.

# Running Code in Interactive Cells

Execute code snippets in individual cells, allowing for step-by-step debugging and variable inspection.

# Displaying Outputs and Visualizations

Use print statements, plots, and markdown cells to display results and integrate visualizations directly in the notebook.
